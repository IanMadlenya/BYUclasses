## Estimating Returns

```python
import pandas as pd
from pandas.io.data import DataReader, get_data_yahoo
import matplotlib.pyplot as plt
```

```python
# Define starting and ending dates for data collection
start = '1/1/1990'
end = '12/31/2012'
```

```python
# Grab S&P500, VIX, ten-year, and three-month T-bill rates
sp500 =get_data_yahoo('^GSPC', start=start, end=end)['Adj Close']
sp500.name = 'SP500'

vix =get_data_yahoo('^VIX', start=start, end=end)['Adj Close']
vix.name = 'VIX'

ten_yr = DataReader('DGS10', 'fred', start=start)

three_mon = DataReader('DGS3MO', 'fred', start=start)
```

```python
# Do some cleaning and combining
ten_yr = ten_yr.ix[ten_yr.DGS10.str.count(r'^\.') != 1].astype(float)
three_mon = three_mon.ix[three_mon.DGS3MO.str.count(r'^\.') != 1].astype(float)

data = ten_yr.join(three_mon)
data = data.join(sp500)
data = data.join(vix)

# Drop non-like observations (obs on different days)
data = data.dropna()
```

```python
# Show a bit of the data
data.head(10)
```

```python
# See how many days worth of data we have for a few years
print(data.ix['2012'].shape[0])
print(data.ix['2002'].shape[0])
print(data.ix['1992'].shape[0])
print(data.ix['1995'].shape[0])
```

It looks like 250 is a good average value. I will then find the percent change for each data point starting with the current one and going back 250 observations.

This will estimate the one-year returns for each day. 

```python
# Pull out the stocks. This doesn't make sense for the T-bill rates
stocks = data[['SP500', 'VIX']]

# Calculate percent changes 250 days apart
changes = stocks.pct_change(periods=250)

# Resample annually by taking the mean for each year
ann_returns = changes.resample('A', how='mean') * 100

# Pull out T-bill data
tbills = data[['DGS10', 'DGS3MO']]

# No need to do pct_changes. Just resample annualy by averaging 
t_bill_returns = tbills.resample('A', how='mean')

ann_returns = ann_returns.join(t_bill_returns)

ann_returns.columns.name= 'Returns (%)'

# Show it off
ann_returns.plot(figsize=(15, 8), linewidth=2.5, title='Estimated Annual returns')
ann_returns.dropna()
```

These results seem a bit to volatile. For example, saying that the S&P 500 returned 30.17% in 1997 seems a bit high.

However, it is encouraging that my simple esimator caught the dotcom bubble in the early 2000's and the housing bubble in 2008-2009.
