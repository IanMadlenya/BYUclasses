```python
import pandas as pd
import numpy as np
from pandas.io.data import get_data_yahoo
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
# Get data
aapl = get_data_yahoo('AAPL')['Adj Close']
```

```python
# Plot the data
aapl.plot()
plt.title('Apple Inc. Stock Prices (Feb 2010 - March 2013)')
plt.ylabel('Price ($)')
plt.xlabel('Date')
```

```python
# Plot the differences
aapl.diff(1).plot()
plt.title('Apple Inc. Stock Price L(1) Differences (2/1/10 - 3/25/13)')
plt.ylabel('Price ($)')
plt.xlabel('Date')
```

```python
# Plot the log
np.log(aapl).plot()
plt.title('Apple Inc. Stock Price (log) (2/1/10 - 3/25/13)')
plt.ylabel('Price ($)')
plt.xlabel('Date')
```

```python
# Plot the log differenced
np.log(aapl).diff(1).plot()
plt.title('Apple Inc. Stock Price (log, L1) (2/1/10 - 3/25/13)')
plt.ylabel('Price ($)')
plt.xlabel('Date')
```

```python
# Get auto correlation and partial autocorrelation coefficients
auto_corr = sm.tsa.stattools.acf(aapl.values)
partial_auto_corr = sm.tsa.stattools.pacf(aapl.values)

# Plot the partials
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(partial_auto_corr, label='Partials')

# Plot the autocorrleation 
ax2 = ax1.twinx()
ax2.plot(auto_corr, 'r', label='Autocor coeffs.',)

ax1.legend(loc=0)
ax2.legend(loc=9)
plt.show()
```
